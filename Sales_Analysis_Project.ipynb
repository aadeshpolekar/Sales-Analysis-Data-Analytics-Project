```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.dropna()
```

```python
print("Total Sales:", df['Sales'].sum())
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')
plt.title("Top 10 Products")
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Category Sales Distribution")
plt.ylabel("")
plt.show()sub_category_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

sub_category_sales.plot(kind='bar')
plt.title("Sub-Category Sales")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.show()
```

```python
sub_category_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

sub_category_sales.plot(kind='bar')
plt.title("Sub-Category Sales")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.show()
```

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_customers.plot(kind='bar')
plt.title("Top 10 Customers")
plt.show()
```

```python
ship_mode_sales = df.groupby('Ship Mode')['Sales'].sum()

ship_mode_sales.plot(kind='bar')
plt.title("Sales by Ship Mode")
plt.show()
```

```python
top_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)

top_cities.plot(kind='bar')
plt.title("Top Cities by Sales")
plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

segment_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Customer Segment Distribution")
plt.show()
```

```python
avg_order = df['Sales'].mean()
print("Average Order Value:", avg_order)
```

```python
region_category = df.groupby(['Region','Category'])['Sales'].sum().unstack()

region_category.plot(kind='bar', stacked=True)
plt.title("Region vs Category Sales")
plt.show()
```

```python
low_products = df.groupby('Product Name')['Sales'].sum().sort_values().head(10)

low_products.plot(kind='bar')
plt.title("Lowest Selling Products")
plt.show()
```
